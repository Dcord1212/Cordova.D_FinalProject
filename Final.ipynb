```python
# Packages 
import pandas as pd
from datetime import datetime
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load in the Data 
#wresteled with weird formatting of old excel file type / excess data. Despite different attempts at using "header count" type... 
#modifiers, reading in as filepath, etc etc. 
#this is a final, cleaned up version where the number of individual variables was greatly reduced.
#This reduction was done using excel
#origionally, tried to load up each of these as a seperate excel sheet loaded up. Online, I looked at working with excel sheets, and... 
#was abble to use sheet_name= to select for each sheet in an excel file
df_og = pd.read_excel('Final Project Correct Data.xlsx', sheet_name='Origional', header = 1, usecols=['Year','Species Name','Number Of Fish (estimated)','Vessel Count','Effort'])
df_Al = pd.read_excel('Final Project Correct Data.xlsx', sheet_name='Al', usecols=['Year','Species Name','Number Of Fish (estimated)','Vessel Count','Effort'])
df_Ko = pd.read_excel('Final Project Correct Data.xlsx', sheet_name='Ko', usecols=['Year','Species Name','Number Of Fish (estimated)','Vessel Count','Effort'])
df_Cook = pd.read_excel('Final Project Correct Data.xlsx', sheet_name='Cook', usecols=['Year','Species Name','Number Of Fish (estimated)','Vessel Count','Effort'])
df_Yak = pd.read_excel('Final Project Correct Data.xlsx', sheet_name='Yak', usecols=['Year','Species Name','Number Of Fish (estimated)','Vessel Count','Effort'])
df_Bri = pd.read_excel('Final Project Correct Data.xlsx', sheet_name='Bri', usecols=['Year','Species Name','Number Of Fish (estimated)','Vessel Count','Effort'])
```

```python
# Plot 1: Fish per area 
fig,axs = plt.subplots(figsize = (20,12))
#set the axis
#in order to avoid personal confusion, did set y# as different variables. While not strictly necessary, helped me wrap my brain around it 
x = df_Al['Year'] # origionally didn't work with df_og, because i miscalculated the file size. Just ended up using year values from one of the other dataframes instead
y1 = df_Al['Number Of Fish (estimated)']
y2 = df_Cook['Number Of Fish (estimated)']
y3 = df_Ko['Number Of Fish (estimated)']
y4 = df_Yak['Number Of Fish (estimated)']
y5 = df_Bri['Number Of Fish (estimated)']
#setting up the plots
axs.plot(x,y1, label= 'Aleutian Islands', color='darkmagenta') 
axs.plot(x,y2, label = 'Kodiak Island', color='magenta')
axs.plot(x,y3,label='Cook Inlet',color='thistle')
axs.plot(x,y4,label='Yakutat',color='crimson')
axs.plot(x,y5,label='Bristol Bay',color='mediumblue')

##Touching up the plot
#here, many of these parameters (at the time of their construction) were not instructions known to me, and statology was a great resource
#Getting the correct fontsize so that each is readable in the powerpoint was a significant task
axs.set_xlabel('Year', fontsize=28)
axs.set_ylabel('Estimated number of Fish Caught (by 10 million)', fontsize=28)
axs.set_title('Sockeye catch over time', fontsize=32)
fig.legend(fontsize=20)
axs.tick_params(axis='x', labelsize=26)
axs.tick_params(axis='y', labelsize=26)
#adding a trendline
#well, I tried to, but then i realized that the trendline was misleading/not super visible anyways due to distribution of graph
#z = np.polyfit(x,y1,1)
#p=np.poly1d(z)
#plt.plot(x,p(x))
```

```python
# Plot 2: Vessel per area
fig,axs = plt.subplots(figsize = (20,12))
#set the axis
x = df_Al['Year']
y1 = df_Al['Vessel Count']
y2 = df_Cook['Vessel Count']
y3 = df_Ko['Vessel Count']
y4 = df_Yak['Vessel Count']
y5 = df_Bri['Vessel Count']
#setting up the plots 
axs.plot(x,y1, label= 'Aleutian Islands', color='darkmagenta') 
axs.plot(x,y2, label = 'Kodiak Island', color='magenta')
axs.plot(x,y3,label='Cook Inlet',color='thistle')
axs.plot(x,y4,label='Yakutat',color='crimson')
axs.plot(x,y5,label='Bristol Bay',color='mediumblue')

##Touching up the plot
axs.set_xlabel('Year',fontsize=28)
axs.set_ylabel('Number of Fishing Vessles',fontsize=28)
axs.set_title('Vessel Count over Time',fontsize=30)
axs.tick_params(axis='x', labelsize=24)
axs.tick_params(axis='y', labelsize=24)
fig.legend(fontsize=20)
```

```python
# Plot 3: Vessel per area
fig,axs = plt.subplots(figsize = (20,12))
#set the axis
x = df_Al['Year']
y1 = df_Al['Effort']
y2 = df_Cook['Effort']
y3 = df_Ko['Effort']
y4 = df_Yak['Effort']
y5 = df_Bri['Effort']
#setting up the plots 
axs.plot(x,y1, label= 'Aleutian Islands', color='darkmagenta') 
axs.plot(x,y2, label = 'Kodiak Island', color='magenta')
axs.plot(x,y3,label='Cook Inlet',color='thistle')
axs.plot(x,y4,label='Yakutat',color='crimson')
axs.plot(x,y5,label='Bristol Bay',color='mediumblue')

#Touching up the plot
axs.set_xlabel('Year',fontsize=28)
axs.set_ylabel('Effort (Salmon per Vessel)',fontsize=28)
axs.set_title('Change in Effort over Time',fontsize=30)
fig.legend(fontsize=18)
axs.tick_params(axis='x', labelsize=24)
axs.tick_params(axis='y', labelsize=24)
```
