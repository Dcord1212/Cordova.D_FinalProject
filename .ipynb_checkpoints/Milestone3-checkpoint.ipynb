```python
#Chinook Data
df_al_chinook = pd.read_excel('Aleutian Island Salmon.xlsx', sheet_name='Chinook', header = 1, usecols=['Year','Area','Species Name','Number Of Fish (estimated)','Landed Weight (lbs)','Whole Weight (lbs)','Vessel Count'])

#display(df_al_chinook)
```

```python
#Chum Data
df_al_chum = pd.read_excel('Aleutian Island Salmon.xlsx', sheet_name='Chum', header = 1, usecols=['Year','Area','Species Name','Number Of Fish (estimated)','Landed Weight (lbs)','Whole Weight (lbs)','Vessel Count'])

#display(df_al_chum)
```

```python
#Sockeye Data
df_al_sockeye = pd.read_excel('Aleutian Island Salmon.xlsx', sheet_name='Sockeye', header = 1, usecols=['Year','Area','Species Name','Number Of Fish (estimated)','Landed Weight (lbs)','Whole Weight (lbs)','Vessel Count'])

#display(df_al_sockeye)
```

```python
import matplotlib.pyplot as plt

fig,axs = plt.subplots(figsize = (10,6))
#first plot temperature data 
##now use axs to label the axis - actual stuff we can manipulate 
#set x and y
x = df_al_pink['Year']
y1 = df_al_pink['Whole Weight (lbs)']
y2 = df_al_coho['Whole Weight (lbs)']
y3 = df_al_sockeye['Whole Weight (lbs)']
y4 = df_al_chinook['Whole Weight (lbs)']
y5 = df_al_chum['Whole Weight (lbs)']
axs.plot(x,y1,marker='.', markerfacecolor = 'white', label = 'Pink', color= 'pink') #plotting number of fish
axs.plot(x,y2,marker='.', label = 'Coho', color = 'green')
axs.plot(x,y3,label='Sockeye')
axs.plot(x,y4,label='Chinook')
axs.plot(x,y5,label='Chum')
axs.set_title = 'Salmon catch over time'
#we could add labels, legend, second y axis label, 
axs.set_xlabel('Year')
axs.set_ylabel('Total mass of fish caught (10 million lbs)')

#now make a legend
fig.legend()
```
